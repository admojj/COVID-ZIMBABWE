# This notebook is for preparing Data for Social Determinants for Health in Africa( The Causes of the Causes)

# First we read the Unemployment Data(Rates)

```python
# import pandas 
import pandas as pd
#read the unemployment data
unempdf1 = pd.read_csv('Data/Africa/Unemployment_Data.csv')
unempdf1
```

```python
unempdf = unempdf1.melt(id_vars=['Country Name', 'Country Code'])
unempdf
```

```python
unempdf.rename(columns={'variable':'year', 'value':'unemp_rate'},inplace=True)
unempdf
```

```python
unempdf.to_pickle('Data/Africa/unempdf_pickle.pkl')
```
