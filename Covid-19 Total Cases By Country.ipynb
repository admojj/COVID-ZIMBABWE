# Section 2 of Notebook - Demonstrate Covid-19 Cases and Deaths in Africa
In this section we will demonstrate steps to show Covid-19 cases and Deaths in Africa
1.Import Africa Shapefile and create a geodataframe
2.Convert df dataframe to a geodataframe
3.Merge the two geodataframes
4.Plot Folium maps showing cases and deaths

# Import Africa Shapefile
To enhance the value of the data we are going to merge a geo dataframe and the df so as to be able to relate cases to place more accurately

```python
import geopandas as gpd
```

```python
#create africa shapefile and assign it to variable africa_gdf
africa_gdf = gpd.read_file('Data/Africa_World_PROD_1_Lsib_2017_Mar.shp')
```

```python
#show africa geodataframe
africa_gdf
```

```python
#plot africa geodataframe
africa_gdf.plot(figsize=(10,10))
```

# Convert df dataframe to a geodataframe to enable merging with africa_gdf geodataframe

```python
# create geodataframe from df and assign it to variable gdf
gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.Long_, df.Lat))
gdf
```

```python
#Check the coordinate reference system for the africa_gdf geodataframe
africa_gdf.crs
```

```python
# project the derived geodataframe gdf to the same CRS as africa_gdf to enable merge
gdf.crs = 'epsg:4326'
gdf.crs
```

```python
# merge the two geodataframe using spatial join sjoin and assign the new geodataframe to gdf2 variable
gdf2 = gpd.sjoin(gdf, africa_gdf, how='inner', op='intersects')
gdf2
```

# Plot a map of Confirmed Covid-19 Cases in Africa
This visualization shows the confirmed Covid-19 cases in Africa
The code imports folium and then applies cartodbpositron for amplified cognition.Popup messages contain location and number of confirmed cases

```python
import folium 
folium.__version__

map3 = folium.Map(location=Zimbabwe_Coordinates, zoom_start=3.2)

folium.TileLayer('cartodbpositron').add_to(map3)


for index, row in gdf2.iterrows():
    country = row['Country_Region']
    lat = row['Lat']
    lon = row['Long_']
    Confirmed = row['Confirmed']
    popup_string = '<b>'+country.upper() +'</b> <br/>Location: '+'('+str(lat)+\
                     ', '+str(lon)+')'+\
                     '<br/>'+\
                    'Confirmed: '+ str(Confirmed)
    popup = folium.Popup(popup_string)
    folium.CircleMarker(
                    location=[lat,lon], \
                    radius=Confirmed / 100 , \
                    color='black', \
                    weight=1, \
                    fill=True, \
                    popup=popup, \
                    fill_color='red', \
                    fill_opacity=1).add_to(map3)
map3
```

# Plotting a Map of the Covid-19 Deaths in Africa
this visualization shows the Covid Deaths recorded in Africa.
First we create a derived geodataframe gdf3 which which contains countries with one or more deaths
then we plot a map, map4 using gdf3.gdf3 is derived from gdf2

```python
gdf3 = gdf2[(gdf2.Deaths != 0)]
gdf3
```

```python
import folium 
folium.__version__

map4 = folium.Map(location=Zimbabwe_Coordinates, zoom_start=3.2)

folium.TileLayer('cartodbpositron').add_to(map4)


for index, row in gdf3.iterrows():
    country = row['Country_Region']
    lat = row['Lat']
    lon = row['Long_']
    Confirmed = row['Deaths']
    popup_string = country + ' Location: '+'('+str(lat)+\
                     ', '+str(lon)+')'+\
                     '<br/>'+\
                    'Deaths: '+ str(Confirmed)
    popup = folium.Popup(popup_string)
    folium.CircleMarker(
                    location=[lat,lon], \
                    radius=Confirmed / 5 , \
                    color='black', \
                    weight=1, \
                    fill=True, \
                    popup=popup, \
                    fill_color='red', \
                    fill_opacity=1).add_to(map4)
map4
```

# Plot a Choropleth Map 

```python
gdf4 = gpd.sjoin(gdf, africa_gdf, how='inner', op='within')
gdf4
```

```python
import folium    

# Picked location Zimbabwe
map5 = folium.Map(location=[-19.015438 , 29.154857],zoom_start=5,tiles="cartodbpositron")

for index, row in africa_gdf.iterrows():
    
    geojson = folium.GeoJson(row['geometry'])
    geojson.add_to(map5)
    
map5
```

```python
import pandas as pd


Zimbabwe_Coordinates = (-19.015438 , 29.154857)

map6 = folium.Map(location=Zimbabwe_Coordinates, zoom_start=3.2)

folium.Choropleth(
    geo_data=gdf4,
    name='choropleth',
    data=gdf4,
    columns=['OBJECTID', 'Deaths'],
    key_on="feature.OBJECTID",
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Africa Covid Cases'
).add_to(map6)

folium.LayerControl().add_to(map6)

map6
```

```python
mymap = folium.Map(location=Zimbabwe_Coordinates, zoom_start=3.2)
myscale = (gdf4['Deaths'].quantile((0.1, 0.25, 0.5, 0.75, 1))).tolist()
folium.Choropleth(
 geo_data=gdf4,
 name='Choropleth',
 data=gdf4,
 columns=['OBJECTID','Deaths'],
 key_on="feature.properties.OBJECTID",
 fill_color='YlGnBu',
 threshold_scale=myscale,
 fill_opacity=1,
 line_opacity=0.2,
 legend_name='Africa',
 smooth_factor=0
)
mymap
```
