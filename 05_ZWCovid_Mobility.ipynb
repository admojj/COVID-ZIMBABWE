```python
import pandas as pd
```

```python
# read the mobility data csv file
zimmobdf = pd.read_csv('Data/Zimbabwe/Global_Mobility_Report.csv')
zimmobdf
```

```python
#rename the column headings
zimmobdf.columns = ['Country Code', 'CNTRY_NAME', 'region1', 'region2', 'iso3', 'FIPS', 'date', 'retail_and_recreation', 'grocery_and_pharmacy', 'parks', 'transit_stations', 'workplaces', 'residential']
zimmobdf
```

```python
#drop columns region1  and region2 
zimmobdf.drop(['region2', 'iso3', 'FIPS'],axis=1, inplace=True)
zimmobdf
```

```python
# Select df.desired column row == Zimbabwe
zimmobility = zimmobdf[(zimmobdf.CNTRY_NAME == "Zimbabwe")]
zimmobility
```

```python
# Select df.desired column row == Zimbabwe
zimmobility2 = zimmobdf[(zimmobdf.region1 == "Harare Province")]
zimmobility2
```

```python
# Using plotly.express
import plotly.express as px

import pandas as pd


fig = px.line(zimmobility2, x='date', y='retail_and_recreation', range_x=['2020-02-15','2020-07-12'], title='Harare retail and recreation mobility trends',)

fig.show()
```

```python
# Using plotly.express
import plotly.express as px

import pandas as pd


fig = px.line(zimmobility2, x='date', y='workplaces', range_x=['2020-02-15','2020-07-12'], title='Harare Workplaces',)

fig.show()
```

```python
# Using plotly.express
import plotly.express as px

import pandas as pd


fig = px.line(zimmobility2, x='date', y='residential', range_x=['2020-02-15','2020-06-28'], title='Harare residential',)

fig.show()
```

```python
import plotly.graph_objects as go
import datetime
import numpy as np
np.random.seed(1)

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
min_date=zimmobility.min()
max_date=zimmobility.max()
```

```python
min_date
```

```python
fig = go.Figure(data=go.Heatmap(
        z=zimmobility['workplaces'],
        x=zimmobility['date'],
        y=zimmobility['region1'],       
        colorscale='plasma',
        colorbar=dict(thickness=20,lenmode='fraction',len=0.8,
                      tickcolor='black',tickfont=dict(size=10),y=0.8,
                     title='mobility workplaces')
))

fig.update_layout(
    #title='Heat Map, Mobility in Zimbabwe, '+min_date+' to '+max_date, 
    titlefont=dict(size=14),
    yaxis_nticks=len(zimmobility['region1'].unique()),
    xaxis_nticks=len(zimmobility['date'].unique()),
    width=1000,
    height=700,
    yaxis=dict(title_text="Province", titlefont=dict(size=12),tickfont=dict(size=9)),
    xaxis=dict(title_text="Report Date",titlefont=dict(size=12),tickfont=dict(size=9)),
    margin=dict(l=50, r=50, b=100, t=100,pad=4)
)

fig.show()
```
