# First we will import the Confirmed Cases by Province in Zimbabwe
Zimbabwe has a total of 10 Provinces in 2020

```python
#import pandas
import pandas as pd
# read the Confirmed cases file and create a dataframe and assign it to the variable zdf
zdf = pd.read_csv('Data/Zimbabwe/Zimbabwe_Covid_Cases.csv')
# show the zdf dataframe
zdf
```

```python
# print the class of the dataframe and the datatypes in the dataframe
zdf.info()
```

```python
zdf.to_pickle('Data/Zimbabwe/zdf_pickle.pkl')
```
